```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read dataset
df = pd.read_csv('cleaned_star_data.csv')
# display first few rows
df.head()
```

```python
# check null values
df.isnull().sum()
```

```python
# drop null values and call it df_cleaned
df_cleaned = df.dropna()
```

```python
# check null values in new cleaned data
df_cleaned.isnull().sum()
```

```python
# plot size
plt.figure(figsize=(5,3))
# seaborn create scatterplot
sns.scatterplot(data=df, x= 'Temperature (K)', y = 'Luminosity(L/Lo)', hue = 'Star type')
# plot titles and labels
plt.title('Hertzsprung–Russell Diagram', fontsize=16)
plt.xlabel('Temperature (K)', fontsize=12)
plt.ylabel('Luminosity(L/Lo)', fontsize=12)
# change labels to log
plt.xscale('log')
plt.yscale('log') 
# tight layout for readability
plt.tight_layout()
#show plot
plt.show()
           
```

```python
# plot size
plt.figure(figsize=(5,3))
# seaborn create bar chart
sns.countplot(data=df, x='Star color')
# plot titles and labels
plt.title('Star Colour by Count')
plt.xlabel('Star color')
plt.ylabel('Count')
# rotate labels on x by 90 degrees to make them readable
plt.xticks(rotation=90)
# show plot
plt.show()
```

```python
# checking star colour counts and finding out why there is a gap
star_counts = df['Star color'].value_counts()
print(star_counts)
```

```python
df['Star color'] = df['Star color'].str.strip()
df['Star color'] = df['Star color'].replace('', np.nan)
df.dropna(subset=['Star color'], inplace=True)
star_counts_cleaned = df['Star color'].value_counts()
print(star_counts_cleaned)
```

```python
# plot size
plt.figure(figsize=(5,3))
all_colours = ['Red', 'Blue-White', 'White', 'Yellow-White', 'Blue']
star_counts = df['Star color'].value_counts()
complete_counts = star_counts.reindex(all_colors, fill_value=0)
# seaborn create bar chart
sns.barplot(x=complete_counts.index, y=complete_counts.values)
# plot titles and labels
plt.title('Star Colour by Count')
plt.xlabel('Star color')
plt.ylabel('Count')
# rotate labels on x by 90 degrees to make them readable
plt.xticks(rotation=90)
# show plot
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(data=df, x= 'Temperature (K)', y = 'Radius(R/Ro)')
plt.title('Temp against Radius aka another hertzsprung russell diagram')
plt.xlabel('Temperature (K)')
plt.ylabel('Radius (R/Ro)')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
